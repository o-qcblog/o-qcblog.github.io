```python
import random
import ipywidgets as widgets
from IPython.display import display

# Generate a random number
number = random.randint(1, 10000)
count = 0

# Create widgets
instruction = widgets.HTML(value="<b>Guess the number game!</b> <br> Enter your guess in the format: <code>&lt; 500</code>, <code>&gt; 600</code>, or <code>= 700</code>")
text_input = widgets.Text(placeholder="Enter your guess")
output = widgets.Output()
submit_button = widgets.Button(description="Submit")

# Define callback function for the button
def on_submit(change):
    global count
    guess_data = text_input.value.strip()
    text_input.value = ""  # Clear input box
    
    if not guess_data:
        return  # Ignore empty input
    
    with output:
        try:
            question = guess_data[0]
            guess = int(guess_data[1:].strip())
            count += 1
            
            if question == '=':
                if guess == number:
                    print(f"üéâ Well done! You guessed it right in {count} steps!")
                    submit_button.disabled = True  # Disable further inputs
                else:
                    print("‚ùå It was a wrong guess :(")
            
            elif question == '>':
                if number >= guess:
                    print(f"‚úÖ Yes! The number is greater than {guess}")
                else:
                    print(f"‚ùå No! The number is not greater than {guess}")
            
            elif question == '<':
                if number <= guess:
                    print(f"‚úÖ Yes! The number is lesser than {guess}")
                else:
                    print(f"‚ùå No! The number is not lesser than {guess}")
            
            else:
                print("‚ö†Ô∏è Invalid entry. Use <, >, or = followed by a number. Example: >40")
        
        except (ValueError, IndexError):
            print("‚ö†Ô∏è Invalid format. Make sure your input follows the format <40, >50, or =60.")

# Connect the button to the callback
submit_button.on_click(on_submit)

# Display the widgets
display(instruction, text_input, submit_button, output)
```
