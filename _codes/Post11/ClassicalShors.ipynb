## Order Finding Classically

```python
#Choose various N and a values

N= 15
a = 18
```

```python
import math
math.gcd(a,N)
```

```python
import matplotlib.pyplot as plotter

z = list(range(N))
y= [ a**z0 % N for z0 in z]

plotter.plot(z,y)
plotter.xlabel('z')
plotter.ylabel(f'{a}^z (mod {N} )')
plotter.show()
```
