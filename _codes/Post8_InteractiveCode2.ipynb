```python
import random
import math
from ipywidgets import Button, Output, VBox, Label, Dropdown
from IPython.display import display

# Initial Variables
count = 0
start = 0
stop = 10000
guess = random.randint(start, stop)
question = ""
output = Output()

# Widgets
question_label = Label(value=f"Is the number {question} {guess}?")
response_dropdown = Dropdown(
    options=["Choose...", "Yes", "No"],
    value="Choose...",
    description="Response:"
)
submit_button = Button(description="Submit")
restart_button = Button(description="Restart", button_style="danger")

# Functions
def restart_game(_):
    """Restart the game with initial conditions."""
    global count, start, stop, guess, question
    count = 0
    start = 0
    stop = 10000
    guess = random.randint(start, stop)
    question = random.choice([">", "<", "="])
    question_label.value = f"Is the number {question} {guess}?"
    response_dropdown.value = "Choose..."
    output.clear_output()

def process_response(_):
    """Process the user's response to the computer's question."""
    global count, start, stop, guess, question
    response = response_dropdown.value

    if response == "Choose...":
        with output:
            print("Please select a valid response: Yes or No.")
        return

    # Increment the count
    count += 1

    # Handle the user's response
    if question == "=" and response == "Yes":
        with output:
            if count < math.log2(10000):
                print(f"I guessed your number in {count} steps! I'm great at this!")
            else:
                print(f"That took me {count} steps. I need to improve!")
        return
    elif question == "=" and response == "No":
        with output:
            print("Hmm, I seem to have made a mistake. Let me try again.")

    # Adjust the range based on the user's feedback
    if question == ">":
        if response == "Yes":
            start = guess
        else:
            stop = guess
    elif question == "<":
        if response == "Yes":
            stop = guess
        else:
            start = guess

    # Check for invalid range
    if start > stop:
        with output:
            print("It seems there's an inconsistency in your responses!")
        return

    # Generate a new guess and question
    guess = random.randint(start, stop)
    question = random.choice([">", "<", "="])
    while ((question == "<" and guess == 0) or (question == ">" and guess == 10000) or (question == ">" and guess == 0) or (question == "<" and guess == 10000)):
        guess = random.randint(start, stop)
        question = random.choice([">", "<", "="])
    question_label.value = f"Is the number {question} {guess}?"
    response_dropdown.value = "Choose..."

# Event Listeners
submit_button.on_click(process_response)
restart_button.on_click(restart_game)

# Display the widgets
controls = VBox([question_label, response_dropdown, submit_button, restart_button, output])
display(controls)

# Initialize the game
restart_game(None)
```
